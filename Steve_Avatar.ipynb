```python
# =============================================
# STEP 1: INSTALL DEPENDENCIES (CPU ONLY)
# =============================================
!apt update -qq
!apt install -y ffmpeg git python3.9 > /dev/null
!pip install -q gtts opencv-python torch==1.13.1 torchvision==0.14.1 numpy==1.21.6 librosa==0.9.2

# =============================================
# STEP 2: SETUP Wav2Lip (CPU VERSION)
# =============================================
import os

# Clean previous installs
!rm -rf Wav2Lip
!git clone https://github.com/Rudrabha/Wav2Lip.git --quiet
os.chdir("Wav2Lip")

# Download pretrained models
!mkdir -p checkpoint
!wget https://github.com/justinjohn0306/Wav2Lip/releases/download/models/wav2lip.pth -O checkpoint/wav2lip.pth -q
!wget https://github.com/justinjohn0306/Wav2Lip/releases/download/models/wav2lip_gan.pth -O checkpoint/wav2lip_gan.pth -q
!wget https://github.com/justinjohn0306/Wav2Lip/releases/download/models/mobilenet.pth -O checkpoint/mobilenet.pth -q

# Patch for CPU inference and librosa compatibility
def patch_code():
    # Force CPU usage
    with open("inference.py", "r") as f:
        code = f.read()
    code = code.replace("torch.cuda.is_available()", "False")
    with open("inference.py", "w") as f:
        f.write(code)

    # Fix librosa melspectrogram call
    with open("audio.py", "r") as f:
        code = f.read()
    code = code.replace("librosa.filters.mel(hp.sample_rate, hp.n_fft, n_mels=hp.num_mels,",
                       "librosa.filters.mel(sr=hp.sample_rate, n_fft=hp.n_fft, n_mels=hp.num_mels,")
    with open("audio.py", "w") as f:
        f.write(code)
patch_code()

# =============================================
# STEP 3: LIP-SYNC EXECUTION (YOUR FILES)
# =============================================
# Define paths (use your uploaded files)
AUDIO_PATH = "/content/Apple.wav"  # Your .wav file
IMAGE_PATH = "/content/Bill.jpg"    # Your avatar image
OUTPUT_VIDEO = "result.mp4"          # Output video

# Verify files exist
if not os.path.exists(AUDIO_PATH):
    raise FileNotFoundError(f"Audio file not found at {AUDIO_PATH}")
if not os.path.exists(IMAGE_PATH):
    raise FileNotFoundError(f"Image file not found at {IMAGE_PATH}")

# Run inference (CPU)
!python inference.py --checkpoint_path checkpoint/wav2lip_gan.pth --face "{IMAGE_PATH}" --audio "{AUDIO_PATH}" --outfile "{OUTPUT_VIDEO}" --static True --pads 0 0 0 0

# =============================================
# STEP 4: PLAY RESULTS (IN COLAB)
# =============================================
from IPython.display import HTML
from base64 import b64encode

def show_video(video_path):
    if not os.path.exists(video_path):
        raise FileNotFoundError(f"Output video not found at {video_path}")
    mp4 = open(video_path, "rb").read()
    data_url = "data:video/mp4;base64," + b64encode(mp4).decode()
    return HTML(f"""
    <video width="70%" controls>
        <source src="{data_url}" type="video/mp4">
    </video>
    """)

print("✨ Lip-Sync Successful! Output:")
show_video(OUTPUT_VIDEO)
```
